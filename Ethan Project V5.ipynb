```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import seaborn as sns
import re

URA_Data = pd.read_csv('Vacant site by URA - additional details.csv')

def clean_and_convert_price(price):
    if isinstance(price, str):
        price = re.sub(r'[^\d.]', '', price)
    try:
        return float(price)
    except ValueError:
        return None
    
URA_Data["psm_per_gpr_or_gfa"] = URA_Data["psm_per_gpr_or_gfa"].apply(clean_and_convert_price)
    
average_prices = URA_Data.groupby(["planning_area", "type_of_region"])["psm_per_gpr_or_gfa"].mean().reset_index()

average_prices = average_prices.sort_values(by="type_of_region")

# Calculate average P/Sqm for each type of region
average_per_type = average_prices.groupby("type_of_region")["psm_per_gpr_or_gfa"].mean()

plt.figure(figsize=(12, 5))
# Define specific colors for each region
color_dict = {'Core Central Region (CCR)': 'blue', 'Outside Core Central Region (OCR)': 'black', 'Rest of Central Region (RCR)': 'lightcoral'}

sns.barplot(data=average_prices, x="planning_area", y="psm_per_gpr_or_gfa", hue="type_of_region", palette="Set3")

# Plot average lines with matching colors
for region, color in color_dict.items():
    avg_color = color
    avg_value = average_per_type[region]
    plt.axhline(y=avg_value, color=avg_color, linestyle='--', label=f'Avg {region}')

plt.title('Average P/Sqm by Planning Area by Type of region', fontsize=10)
plt.xlabel('Planning Area')
plt.ylabel('P/Sqm')
plt.xticks(rotation=75, fontsize=8)
plt.legend()
plt.show()
```
