```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv("VacantSitesSoldByURA.csv")
df_counts = df.groupby(['type_of_devt_code', 'no_of_bids']).size().reset_index(name='counts')


# Sample frequency data (you should replace this with your own data)
frequencies = df.groupby(['no_of_bids', 'type_of_devt_code']).size().reset_index(name='counts')

# Scale the frequencies to determine circle sizes
min_size = 10
max_size = 100
scaled_sizes = [min_size + (max_size - min_size) * (freq - min(frequencies)) / (max(frequencies) - min(frequencies)) for freq in frequencies]

# Define circle coordinates
x_coordinates = 'type_of_devt_code'
y_coordinates = 'no_of_bids'

# Create a scatter plot with circles
plt.figure(figsize=(8, 4))
plt.scatter(x_coordinates, y_coordinates, s=scaled_sizes, c='blue', alpha=0.7)
plt.title("Circles with Sizes Based on Frequency")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")


for i, freq in enumerate(frequencies):
    plt.annotate(f"Freq: {freq}", (x_coordinates[i], y_coordinates[i]), ha='center', va='center')

# Show the plot
plt.show()
```

```python
# Import Data
df_counts = df.groupby(['no_of_bids', 'type_of_devt_code']).size().reset_index(name='counts')

# Sort the data by 'type_of_devt_code' and 'no_of_bids'
df_counts = df_counts.sort_values(by=['type_of_devt_code', 'no_of_bids'])

# Draw Stripplot
fig, ax = plt.subplots(figsize=(10, 5), dpi=80)    
sns.stripplot(data=df_counts, x='type_of_devt_code', y='no_of_bids', hue='counts', ax=ax)

# Decorations
plt.title('Counts Plot - Size of circle is bigger as more points overlap', fontsize=22)
plt.xticks(rotation=90)
plt.show()
```

```python
# Import Data
df_counts = df.groupby(['no_of_bids', 'type_of_devt_code']).size().reset_index(name='counts')

# Create a scatter plot with larger circles based on 'counts'
fig, ax = plt.subplots(figsize=(10, 5), dpi=80)
scatter = ax.scatter(df_counts['type_of_devt_code'], df_counts['no_of_bids'], s=df_counts['counts'] * 10, c=df_counts['counts'], cmap='viridis', alpha=0.5)

# Add a colorbar for the size of circles
cbar = plt.colorbar(scatter)
cbar.set_label('Counts')

# Decorations
plt.title('Type of Development Code by Number of Bids', fontsize=18)
plt.xlabel('Type of Development Code')
plt.ylabel('Number of Bids')
plt.xticks(rotation=90)

plt.show()
```
