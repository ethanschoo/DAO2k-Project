```python
# Import Data
df = pd.read_csv("VacantSitesSoldByURA.csv")
df_counts = df.groupby(['type_of_devt_code', 'no_of_bids']).size().reset_index(name='counts')

# Draw Stripplot
fig, ax = plt.subplots(figsize=(16,10), dpi= 80)    
sns.stripplot(df_counts.type_of_devt_code, df_counts.no_of_bids, size=df_counts.counts*2, ax=ax)

# Decorations
plt.title('Counts Plot - Size of circle is bigger as more points overlap', fontsize=22)
plt.show()
```

```python
import numpy as np

# Sample frequency data (you should replace this with your own data)
frequencies = df.groupby(['no_of_bids', 'type_of_devt_code']).size().reset_index(name='counts')

# Scale the frequencies to determine circle sizes
min_size = 10
max_size = 100
scaled_sizes = [min_size + (max_size - min_size) * (freq - min(frequencies)) / (max(frequencies) - min(frequencies)) for freq in frequencies]

# Define circle coordinates
x_coordinates = 'type_of_devt_code'
y_coordinates = 'no_of_bids'

# Create a scatter plot with circles
plt.figure(figsize=(8, 4))
plt.scatter(x_coordinates, y_coordinates, s=scaled_sizes, c='blue', alpha=0.7)
plt.title("Circles with Sizes Based on Frequency")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")


for i, freq in enumerate(frequencies):
    plt.annotate(f"Freq: {freq}", (x_coordinates[i], y_coordinates[i]), ha='center', va='center')

# Show the plot
plt.show()
```
